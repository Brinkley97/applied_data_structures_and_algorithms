```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes and edges for the research questions
G.add_edge("What is Graph Theory?", "Graph")
G.add_edge("What is Graph Theory?", "Network")
G.add_edge("What is Graph Theory?", "What are the types?")
G.add_edge("Graph", "Directed")
G.add_edge("Graph", "Undirected")
G.add_edge("What are classifications of graphs?", "Static models")
G.add_edge("What are classifications of graphs?", "Generative models")
G.add_edge("What are classifications of graphs?", "Evolving models")

# Draw the graph
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42)  # Position the nodes
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightblue", font_size=10, font_weight="bold")
plt.title("Research Questions as a Graph", fontsize=16)
plt.show()
```

```python
def generate_rand_graph(N: int):
    """Generate a random graph"""
    g = {}
    domain = list(range(1, N))
    low, high = domain[0], domain[-1] + 1  # Adjust high for inclusivity

    for key in domain:
        print(f"Key: {key}")
        
        # Create a filtered domain excluding the current key
        filtered_domain_list = []
        for value in domain:
            if value != key:
                filtered_domain_list.append(value)

        # Randomize the length of values for the current key, ensuring it's not larger than filtered_domain_list
        max_length = len(filtered_domain_list)
        rand_value_len = np.random.randint(low=1, high=max_length + 1)  # `+1` to include max_length as a possibility

        # Generate a list of unique random values
        rand_values = np.random.choice(filtered_domain_list, size=rand_value_len, replace=False).tolist()
        
        # Assign to the graph dictionary
        g[key] = rand_values

        print(g)
        print()
```

```python
N = 5
generate_rand_graph(N)
```
