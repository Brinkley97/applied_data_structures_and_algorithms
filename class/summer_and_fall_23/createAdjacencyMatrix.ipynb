# Create Adjacency Matrix from Graph

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Create an empty graph
G = nx.Graph()

# Add nodes (5 nodes in this case)
nodes = [0, 1, 2, 3, 4]
G.add_nodes_from(nodes)

# Add edges (5 edges connecting the nodes)
edges = [(0, 1), (1, 2), (2, 3), (3, 4), (4, 0)]
G.add_edges_from(edges)

# Draw the graph
pos = nx.spring_layout(G)  # Positions of nodes for visualization
nx.draw(G, pos, with_labels=True, node_size=500, node_color='lightblue', font_size=12, font_color='black')
plt.title("Simple Graph with 5 Nodes and 5 Edges")
plt.show()
```

```python
def create_adjacency_matrix(edges: list) -> list:
    size_of_adjacency_matrix = len(edges)
    # Initialize an empty adjacency matrix as a list of lists filled with zeros
    adjacency_matrix = [[0] * size_of_adjacency_matrix for _ in range(size_of_adjacency_matrix)]
    print("Original adjacency matrix", adjacency_matrix)
    
    print()
    for edge in edges:
        print(edge)
        l, r = edge
        adjacency_matrix[l][r] = 1
        adjacency_matrix[r][l] = 1
        print(adjacency_matrix)
        print()
        
    return adjacency_matrix
```
